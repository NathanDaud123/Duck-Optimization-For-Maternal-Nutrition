```python
import pandas as pd

url_mp = 'https://github.com/NathanDaud123/Puma/blob/main/Dataset/MP.csv?raw=true'
url_sn = 'https://github.com/NathanDaud123/Puma/blob/main/Dataset/SN.csv?raw=true'
url_sh = 'https://github.com/NathanDaud123/Puma/blob/main/Dataset/SH.csv?raw=true'
url_sy = 'https://github.com/NathanDaud123/Puma/blob/main/Dataset/SY.csv?raw=true'
url_plk = 'https://github.com/NathanDaud123/Puma/blob/main/Dataset/PLK.csv?raw=true'

mp = pd.read_csv(url_mp)
sn = pd.read_csv(url_sn)
sh = pd.read_csv(url_sh)
sy = pd.read_csv(url_sy)
plk = pd.read_csv(url_plk)
```

```python
def hitung_total_karbo(individual):
    total_karbo = 0
    for i in range(0, 15, 5):  # Start from 0, end at 15 with a step of 5
        total_karbo = mp['carbo'][round(individual[i])-1] + sn['carbo'][round(individual[i+1])-1] + sh['carbo'][round(individual[i+2])-1] + sy['carbo'][round(individual[i+3])-1] + plk['carbo'][round(individual[i+4])-1]
        # print(f'Carbo: {mp["carbo"][individual[i]-1]} + {sn["carbo"][individual[i+1]-1]} + {sh["carbo"][individual[i+2]-1]} + {sy["carbo"][individual[i+3]-1]} + {plk["carbo"][individual[i+4]-1]}')

    return total_karbo

def hitung_total_protein(individual):
    total_protein = 0
    for i in range(0, 15, 5):
        total_protein += mp['protein'][round(individual[i])-1] + sn['protein'][round(individual[i+1])-1] + sh['protein'][round(individual[i+2])-1] + sy['protein'][round(individual[i+3])-1] + plk['protein'][round(individual[i+4])-1]
        # print(f'Protein: {mp["protein"][individual[i]-1]} + {sn["protein"][individual[i+1]-1]} + {sh["protein"][individual[i+2]-1]} + {sy["protein"][individual[i+3]-1]} + {plk["protein"][individual[i+4]-1]}')

    return total_protein

def hitung_total_energy(individual):
    total_energy = 0
    for i in range(0, 15, 5):
        total_energy += mp['energy'][round(individual[i])-1] + sn['energy'][round(individual[i+1])-1] + sh['energy'][round(individual[i+2])-1] + sy['energy'][round(individual[i+3])-1] + plk['energy'][round(individual[i+4])-1]
        # print(f'Energy: {mp["energy"][individual[i]-1]} + {sn["energy"][individual[i+1]-1]} + {sh["energy"][individual[i+2]-1]} + {sy["energy"][individual[i+3]-1]} + {plk["energy"][individual[i+4]-1]}')

    return total_energy

def hitung_total_lemak(individual):
    total_fat = 0
    for i in range(0, 15, 5):
        total_fat += mp['fat'][round(individual[i])-1] + sn['fat'][round(individual[i+1])-1] + sh['fat'][round(individual[i+2])-1] + sy['fat'][round(individual[i+3])-1] + plk['fat'][round(individual[i+4])-1]
        # print(f'Fat: {mp["fat"][individual[i]-1]} + {sn["fat"][individual[i+1]-1]} + {sh["fat"][individual[i+2]-1]} + {sy["fat"][individual[i+3]-1]} + {plk["fat"][individual[i+4]-1]}')

    return total_fat

def total_price(individual):
    tot_harga = 0
    for i in range(0, 15, 5):
        tot_harga += mp['price'][round(individual[i])-1] + sn['price'][round(individual[i+1])-1] + sh['price'][round(individual[i+2])-1] + sy['price'][round(individual[i+3])-1] + plk['price'][round(individual[i+4])-1]
        # print(f'Fat: {mp["fat"][individual[i]-1]} + {sn["fat"][individual[i+1]-1]} + {sh["fat"][individual[i+2]-1]} + {sy["fat"][individual[i+3]-1]} + {plk["fat"][individual[i+4]-1]}')
    return tot_harga

def cost_function(individual):
    kebutuhan_energi = 2249.4046
    kebutuhan_karbo = 377.41069
    kebutuhan_protein = 94.3526725
    kebutuhan_lemak = 64.7834611
    total_karbo = hitung_total_karbo(individual)
    total_energi = hitung_total_energy(individual)
    total_protein = hitung_total_protein(individual)
    total_lemak = hitung_total_lemak(individual)

    penalty_karbo = abs(kebutuhan_karbo - total_karbo)
    penalty_energi = abs(kebutuhan_energi - total_energi)
    penalty_protein = abs(kebutuhan_protein - total_protein)
    penalty_lemak = abs(kebutuhan_lemak - total_lemak)
    harga = total_price(individual)

    # print(f'Energy = {total_energi}')
    # print(f'Karbo = {total_karbo}')
    # print(f'Protein = {total_protein}')
    # print(f'Lemak = {total_lemak}')
    # print(f'Harga = {harga}')

    w1 = 0.985
    w2 = 0.015
    
    penalty_nutrition = w1*(penalty_karbo + penalty_energi  + penalty_lemak + penalty_protein)
    penalty_price = w2*harga
    
    # print("nutrisi : ", penalty_nutrition)
    # print("price : ", penalty_price)
    
    fitness = penalty_nutrition + penalty_price
    
    return fitness  # Adjusted to avoid division by zero


def daftar_makanan(best):
    print("="*100)
    print(best)
    print("="*30 + " REKOMENDASI MENU " + "="*30)
    print("Makan Pagi")
    print(f'Makanan pokok\t = {mp["menu"][round(best[0])-1]}')
    print(f'Sumber Nabati\t = {sn["menu"][round(best[1])-1]}')
    print(f'Sumber Hewani\t = {sh["menu"][round(best[2])-1]}')
    print(f'Sayuran\t\t = {sy["menu"][round(best[3])-1]}')
    print(f'Pelengkap\t = {plk["menu"][round(best[4])-1]}')
    print(f'Harga\t = {mp["price"][round(best[0])-1]+sn["price"][round(best[1])-1]+sh["price"][round(best[2])-1]+sy["price"][round(best[3])-1]+plk["price"][round(best[4])-1]}')
    print("Makan Siang")
    print(f'Makanan pokok\t = {mp["menu"][round(best[5])-1]}')
    print(f'Sumber Nabati\t = {sn["menu"][round(best[6])-1]}')
    print(f'Sumber Hewani\t = {sh["menu"][round(best[7])-1]}')
    print(f'Sayuran\t\t = {sy["menu"][round(best[8])-1]}')
    print(f'Pelengkap\t = {plk["menu"][round(best[9])-1]}')
    print(f'Harga\t = {mp["price"][round(best[5])-1]+sn["price"][round(best[6])-1]+sh["price"][round(best[7])-1]+sy["price"][round(best[8])-1]+plk["price"][round(best[9])-1]}')
    print("Makan Malam")
    print(f'Makanan pokok\t = {mp["menu"][round(best[10])-1]}')
    print(f'Sumber Nabati\t = {sn["menu"][round(best[11])-1]}')
    print(f'Sumber Hewani\t = {sh["menu"][round(best[12])-1]}')
    print(f'Sayuran\t\t = {sy["menu"][round(best[13])-1]}')
    print(f'Pelengkap\t = {plk["menu"][round(best[14])-1]}')
    print(f'Harga\t = {mp["price"][round(best[10])-1]+sn["price"][round(best[11])-1]+sh["price"][round(best[12])-1]+sy["price"][round(best[13])-1]+plk["price"][round(best[14])-1]}')
```

```python
import os
import pandas as pd

# Assuming the DataFrames mp, sn, sh, sy, plk are already defined as shown in the initial query

def write_menu_to_csv(best, filename='recommended_menu.csv'):
    # Create an empty list to store menu information
    menu_data = []
    
    # Define meal times for clarity
    meal_times = ['Makan Pagi', 'Makan Siang', 'Makan Malam']
    
    # Loop through each meal time
    for i, meal_time in enumerate(meal_times):
        start_index = i * 5
        # Extract the menu items for the current meal
        meal_items = [
            ('Makanan pokok', mp, best[start_index]),
            ('Sumber Nabati', sn, best[start_index + 1]),
            ('Sumber Hewani', sh, best[start_index + 2]),
            ('Sayuran', sy, best[start_index + 3]),
            ('Pelengkap', plk, best[start_index + 4]),
        ]
        
        # Loop through each menu item and append its details to the menu_data list
        for category, df, index in meal_items:
            menu_data.append({
                'Waktu': meal_time,
                'Kategori': category,
                'Menu': df['menu'][round(index)-1],
                'Karbo': df['carbo'][round(index)-1],
                'Energi': df['energy'][round(index)-1],
                'Lemak': df['fat'][round(index)-1],
                'Protein': df['protein'][round(index)-1],
                'Harga': df['price'][round(index)-1],
            })
    
    # Convert the menu_data list to a DataFrame
    menu_df = pd.DataFrame(menu_data)
    
    # Check if the file exists to determine if headers should be written
    file_exists = os.path.isfile(filename)
    
    # Write the DataFrame to a CSV file in append mode
    menu_df.to_csv(filename, mode='a', index=False, header=not file_exists, encoding='utf-8')
    
# Example usage
# best_individual = [1, 2, 3, 4, 5, 6, 7, 8, 9, 1, 2, 3, 4, 5, 6]  # This should be replaced with the actual best individual
# write_menu_to_csv(best_individual)
```

```python
import numpy as np
import random

convergence_cost = []

class Duck:
    g = None
    gbest = float('+inf')

    # def __init__(self, dimension, xrange, targetFunc):
    def __init__(self, dim, min, max, targetFunc):
        self.d = dim
        self.xmin = min
        self.xmax = max
        self.targetFunc = targetFunc
        self.x = np.random.rand(self.d) * (self.xmax - self.xmin) + self.xmin
        self.pbest = float('+inf')
        self.p = self.x
        self.updateFitness()

    def clipX(self):
        self.x[self.x > self.xmax] = self.xmax
        self.x[self.x < self.xmin] = self.xmin

    def updateFitness(self):
        fitness = self.fitness()
        if fitness < self.pbest:
            self.pbest = fitness
            self.p = self.x
            if self.pbest < Duck.gbest:
                Duck.gbest = self.pbest
                Duck.g = self.p

    def fitness(self, x=None):
        if x is None:
            return self.targetFunc(self.x)
        else:
            x = np.clip(x, 1, 11)  
            return self.targetFunc(x)


class DSA:
    def __init__(self, particleNum, dim, iterTimes, targetFunc,  min, max, P=0.5, FP=0.618):
        Duck.g = None
        Duck.gbest = float('+inf')
        self.particleNum = particleNum
        self.dim = dim
        self.tmax = iterTimes
        self.P = P
        self.FP = FP
        self.particles = [Duck(dim, min, max, targetFunc) for i in range(particleNum)]

    def run(self):
        self.gbestList = []
        indice = [i for i in range(len(self.particles))]
        for t in range(1, self.tmax + 1):
            K = np.sin(2 * np.random.random()) + 1
            Miu = K * (1 - t / self.tmax)
            CF1 = 1 / self.FP * np.random.random()
            CF2 = 1 / self.FP * np.random.random()
            KF1 = 1 / self.FP * np.random.random()
            KF2 = 1 / self.FP * np.random.random()
            # Exploration phase Begin
            for i, particle in enumerate(self.particles):
                if self.P > np.random.random():
                    particle.x = particle.x + Miu * particle.x * np.sign(np.random.random() - 0.5)
                else:
                    remain = list(indice)
                    remain.remove(i)
                    j = random.choice(remain)
                    particle.x = particle.x + CF1 * (Duck.g - particle.x) + CF2 * (self.particles[j].x - particle.x)
            for particle in self.particles:
                particle.clipX()
                particle.updateFitness()
            # Exploration phase End
            # ========================
            # Exploitation phase Begin
            for i, particle in enumerate(self.particles):
                x = particle.x + Miu * (Duck.g - particle.x)
                if particle.fitness(x) < particle.fitness():
                    particle.x = x
                else:
                    remain = list(indice)
                    remain.remove(i)
                    j, k = random.choices(remain, k=2)
                    particle.x = particle.x + KF1 * (Duck.g - particle.x) + KF2 * (
                            self.particles[k].x - self.particles[j].x)
            for particle in self.particles:
                particle.clipX()
                particle.updateFitness()
            self.gbestList.append(Duck.g)
            convergence_cost.append(Duck.gbest)
        return Duck.gbest, Duck.g, list(self.gbestList)

if __name__ == "__main__":
    algorithm = DSA(10, 15, 10, cost_function, 1, 11, P=0.5, FP=0.618)
    gb, x, gbHistory = algorithm.run()
    
    # For list
    # print(convergence_cost)
    unique_values, first_occurrence_indexes = np.unique(convergence_cost, return_index=True)
    last_three_unique_values = unique_values[:3]
    last_three_indexes = first_occurrence_indexes[:3]

    # print(unique_values)
    # print(last_three_unique_values)
    # print(last_three_indexes)
    
    # print('best fitness:', gb)
    # print('best x:', x)
    # print("="*20)
    # print(gbHistory)
    # print(convergence_cost)
    
    for individual in last_three_indexes:
        print("Indeks dari 3 Nilai Unik Terakhir:", individual)
        print("="*100)
        print("Cost Unik Terakhir:", convergence_cost[individual])
        daftar_makanan(gbHistory[individual])
        write_menu_to_csv(gbHistory[individual])

    # return best, best_cost, convergence_cost
```
